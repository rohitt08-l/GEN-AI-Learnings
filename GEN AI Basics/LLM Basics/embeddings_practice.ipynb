```python
# ============================================================
# SAMPLE DATASET
# ============================================================

sentences = [
    ["i", "love", "artificial", "intelligence"],
    ["machine", "learning", "is", "powerful"],
    ["deep", "learning", "uses", "neural", "networks"],
    ["chatgpt", "is", "an", "ai", "assistant"],
    ["natural", "language", "processing", "is", "interesting"],
    ["ai", "is", "transforming", "technology"],
]
```

```python
# ============================================================
# 1. WORD2VEC EMBEDDINGS
# ============================================================

print("\n================ WORD2VEC =================\n")

word2vec_model = Word2Vec(
    sentences,
    vector_size=50,
    window=3,
    min_count=1,
    workers=4
)

print("Vocabulary:")
print(word2vec_model.wv.index_to_key)

print("\nVector for 'ai':")
print(word2vec_model.wv["ai"])

print("\nMost Similar Words to 'learning':")
print(word2vec_model.wv.most_similar("learning"))
```

```python
# ============================================================
# WORD SIMILARITY
# ============================================================

similarity = word2vec_model.wv.similarity("ai", "intelligence")

print("\nSimilarity between 'ai' and 'intelligence':")
print(similarity)



# ============================================================
# 2. FASTTEXT EMBEDDINGS
# ============================================================

print("\n================ FASTTEXT =================\n")

fasttext_model = FastText(
    sentences,
    vector_size=50,
    window=3,
    min_count=1,
    workers=4
)

print("Vector for 'technology':")
print(fasttext_model.wv["technology"])

print("\nFastText can handle unknown words better.")

print("\nVector for unknown word 'technologies':")
print(fasttext_model.wv["technologies"])
```
